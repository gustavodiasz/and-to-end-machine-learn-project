## **4.0 SELECIONE UMA MEDIDA DE DESEMPENHO**

**Seu próximo passo é selecionar uma medida de desempenho.**

**Uma medida típica de desempenho para problemas de regressão é a _Raiz do Erro Quadrático Médio_ (sigla RMSE, em inglês).**


$
\text{RMSE}(\mathbf{X}, h) = \sqrt{\frac{1}{m}\sum\limits_{i=1}^{m}\left(h(\mathbf{x}^{(i)}) - y^{(i)}\right)^2}
$

**_Equation 4-1: Root Mean Square Error (RMSE)_**

**Onde,**

* **$m$ é o número de instâncias no conjunto de dados no qual você está medindo o RMSE;**

* **$\mathbf{x}^{(i)}$  é um vetor de todos os valores da característica _`excluindo o rótulo`_ da _i-ésima_ instância do conjunto de dados;**

* **$\mathbf{y}^{(i)}$ é seu rótulo _`o valor desejado da saída para aquela instância`_ (preço do imóvel);**

* **$\mathbf{x}$ é uma matriz que contém todos os valores da característica _`(é a tabela toda sem a coluna dos preços)`_;**

* **$h$ é a função de previsão do seu sistema. Quando recebe o vetor $\mathbf{x}^{(i)}$, ele exibe um valor previsto $h(x^{(i)}) = \hat{y}^{(i)}$**

**Concluindo:**

**$\text{RMSE}(\mathbf{X}, h)$ `é a função de custo` medida no conjunto de exemplos utilizando sua função de previsão $h(·)$**


--

**Representando a primeira linha de um DataSet em Modelos de ML**

$


$

## **11.0 CUSTOMIZE TRANSFORMADORES**

```python
from sklearn.base import BaseEstimator, TransformerMixin

col_names = 'total_bedrooms', 'population', 'households','total_rooms' #gera uma tupla
bedrooms_ix,  population_ix, households_ix, rooms_ix = [housing.columns.get_loc(c) for c in col_names] 

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room=True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]

attrib_add = CombinedAttributesAdder(add_bedrooms_per_room=True)

tranformed_Values = attrib_add.transform(housing.values)
```

```python
DataFrame_tranformed_Values = pd.DataFrame(tranformed_Values,
                                           columns = list(housing.columns) + ['rooms_per_household', 
                                                                              'population_per_household', 
                                                                              'bedrooms_per_room'],
                                           index=housing.index)

DataFrame_tranformed_Values.head(4)
```

## **12.0 ESCALONAMENTO DAS CARACTERÍSTICAS**

**Resumo:**

O escalonamento de atributos numéricos é essencial para o bom desempenho da maioria dos algoritmos de Aprendizado de Máquina, especialmente quando as variáveis têm escalas muito diferentes. Existem duas técnicas comuns:

1. **Escalonamento Min-Max (normalização)**: Redimensiona os dados para o intervalo \[0, 1], utilizando o mínimo e máximo de cada atributo. É sensível a outliers.

2. **Padronização**: Subtrai a média e divide pelo desvio padrão, resultando em atributos com média 0 e variância 1. É mais robusta a outliers, mas não limita os valores a um intervalo fixo.

O Scikit-Learn oferece os transformadores `MinMaxScaler` e `StandardScaler` para essas tarefas. Importante: o ajuste (`fit`) deve ser feito apenas nos dados de **treinamento**, e depois aplicado (com `transform`) também ao conjunto de **teste** e a novos dados.


```python
from sklearn.preprocessing import StandardScaler

StandardScaler = StandardScaler()
housing_stander = StandardScaler.fit_transform(housing_num)
DataFramehousing_stander = pd.DataFrame(housing_stander,
                                        columns=housing_num.columns,
                                        index=housing_num.index) 
DataFramehousing_stander['median_income'].hist(grid=False,
                                               figsize=(10, 5),
                                               color='darkblue',
                                               bins=20);

media = DataFramehousing_stander['median_income'].mean()

plt.axvline(media, color='red', linestyle='--', linewidth=2, label=f'Média: {media:.2f}')
plt.legend()
plt.show()
```

## **13.0 PIPELINES**

**Como você pode ver, existem muitas etapas de transformação de dados que precisam ser executadas na ordem correta.**

**Felizmente, o Scikit-Learn fornece a classe Pipeline para ajudar com tais sequências de transformações.**

**Eis um pequeno pipeline para os atributos numéricos.**

```python
from sklearn.pipeline import Pipeline

num_pipeline = Pipeline([
                         ('imputer', SimpleImputer(strategy="median")),
                         ('attribs_adder', CombinedAttributesAdder()),
                         ('std_scaler', StandardScaler)
                                                         ])
num_pipeline.fit_transform(housing_num)
```

**Seria muito conveniente se pudéssemos construir um pipeline que lidasse com dados numéricos e categóricos de uma só vez, para evitar o trabalho de sempre ter que separar manualmente as colunas numéricas das categóricas.**

**Felizmente, o Scikit-Learn disponibiliza uma classe chamada `ColumnTransformer` que ajuda a fazer isso.**

```python
from sklearn.compose import ColumnTransformer

columns_num_attribs = list(housing_num)
columns_cat_attribs = ['ocean_proximity']

full_pipeline = ColumnTransformer([
                                   ("num", num_pipeline, columns_num_attribs),
                                   ("cat", OneHotEncoder(), columns_cat_attribs)
                                                                                ])
housing_prepared = full_pipeline.fit_transform(housing)  

name_transformed_Columns=['rooms_per_household', 
                          'population_per_household', 
                          'bedrooms_per_room']

name_columns_categoric = list(full_pipeline.named_transformers_["cat"].get_feature_names_out(["ocean_proximity"]))

Full_DataFrame_transformed=pd.DataFrame(housing_prepared,
                                        columns=columns_num_attribs + name_transformed_Columns + name_columns_categoric)
```

## **14.0 ESCOLHA E TREINE UM MODELO**

### **14.1 Modelo de Regressão Linear**

$\text{Equation 4-1: Linear Regression model prediction}$

$$
\hat{y} = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \dots + \theta_n x_n
$$

$\text{Equation 4-2: Linear Regression model prediction (vectorized form)}$

$$
\hat{y} = h_{\boldsymbol{\theta}}(\mathbf{x}) = \boldsymbol{\theta} \cdot \mathbf{x} = \theta^T \cdot \mathbf{X}
$$



### Método dos Mínimos Quadrados

Para encontrar o valor de $\theta$, que minimiza a função de custo, existe uma solução de forma 
fechada — em outras palavras, uma equação matemática que dá o resultado diretamente. 
Isto é chamado de Método dos Mínimos Quadrados

$\text{Equation 4-4: Método dos Mínimos Quadrados}$

$$
\hat{\boldsymbol{\theta}} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}
$$

* $\hat{\boldsymbol{\theta}}$ $\text{É o valor de}$ $\theta$ $\text{que minimiza a função de custo.}$
* $y$ $\text{É o vetor dos valores do alvo contendo}$ $y^{(1)}$ $a$ $y^{(m)}$
.

**Geraremos alguns dados de aparência linear para testar esta equação**

**Equação de referência:** $$y=4+3x_{i}+\text{ruído}$$

```python
import numpy as np

X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)


plt.plot(X,
         y, 
         "b.")
plt.xlabel("$x_{1}$", 
           fontsize=12)
plt.ylabel("$y$", 
           rotation=0, 
           fontsize=12)

plt.axis([0, 2.4, 0, 15]);
```

**Aplicando a fórmula dos minimos quadrados para determinar $\hat{\boldsymbol{\theta}}$**

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

X_b = np.c_[np.ones((100, 1)), X]  # add x0 = 1 to each instance
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

theta_best
```

**Plotando a função de previsão em meio aos dados usados**

```python
X_new = np.array([[0], [4]])
X_new_b = np.c_[np.ones((2, 1)), X_new]  # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)

plt.plot(X_new, 
         y_predict, 
         "r-")
plt.plot(X, 
         y, 
         "b.")
plt.axis([0, 2.4, 0, 12])
plt.show()
```

### Usando um algorítimo de Regressão linear para prevê o preço dos imóveis.

Relembrando que lá atrás separamos oque era $x$ e oque era $y$ no nosso DataFrame

```python
housing = strat_train_set.drop("median_house_value",
                               axis=1)
housing_labels = strat_train_set["median_house_value"].copy()
```

**Aplicando nosso Pipeline nos dados**

```python
from sklearn.compose import ColumnTransformer

housing_num = housing.drop('ocean_proximity',
                            axis=1)

columns_num_attribs = list(housing_num)
columns_cat_attribs = ['ocean_proximity']

full_pipeline = ColumnTransformer([
                                   ("num", num_pipeline, columns_num_attribs),
                                   ("cat", OneHotEncoder(), columns_cat_attribs)
                                                                                ])
housing_prepared = full_pipeline.fit_transform(housing)  
```

**Treinado o Algoritimo, isto é, Determinando a Matrix $\hat{\boldsymbol{\theta}}$**

```python
from sklearn.linear_model import LinearRegression

LinearRegression = LinearRegression()

LinearRegression.fit(housing_prepared, housing_labels)
```

```python
LinearRegression.intercept_, LinearRegression.coef_
```
